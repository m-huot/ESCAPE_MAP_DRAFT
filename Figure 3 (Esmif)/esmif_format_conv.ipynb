# Convert sequences in generate_seq to esmif format.

```python
import os
from Bio import SeqIO

# Define input and output directories
input_dir = "generated_seq"
output_dir = "seq_esmif_format"
prefix = "TNLCPFGEVFNATRFA"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Process all fasta files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".fasta"):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)

        with open(input_path, "r") as infile, open(output_path, "w") as outfile:
            for record in SeqIO.parse(infile, "fasta"):
                # Modify sequence: prepend prefix and remove last 15 characters
                modified_seq = prefix + str(record.seq)
                
                # Write to output file
                record.seq = modified_seq
                SeqIO.write(record, outfile, "fasta")

print("Processing complete! Modified FASTA files are saved in 'seq_esmif_format'.")
```
